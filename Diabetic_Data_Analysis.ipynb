```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data present in the TSV File
df = pd.read_csv("diabetese.tsv", sep ="\t")
```

```python
df.head()
```

```python
df.info()
```

```python
# let's find the average age 0f diabetic patients
mean_age = df['AGE'].mean()
print(f"Mean age is : {mean_age}")
```

```python
# Variance
var_age = df['AGE'].var()
print(f"Variance of age is : {var_age}")
```

```python
# Standard Deviation
std_age = df['AGE'].std()
print(f"Standard deviation of age is : {std_age}")
```

```python
df.describe()
```

```python
# null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
## There is no null value present in the Diabetic Dataset
```

```python
# Boxplot for the age field
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'],showmeans=True)
plt.grid(color = 'gray', linestyle = "dotted")
plt.xlabel("Age")
plt.ylabel("Years")
plt.show()
```

```python
# boxplotting got BMI
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'],vert=False, showmeans=True)
plt.grid(color ='gray', linestyle = 'dotted')
plt.xlabel("Value")
plt.ylabel("BMI")
plt.show
```

```python
# boxplotting got BP
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'],vert=False, showmeans=True)
plt.grid(color ='gray', linestyle = 'dotted')
plt.xlabel("Value")
plt.ylabel("BP")
plt.show
```

```python
# I want to understand what age of the patients are mostly diabetic


# we can create Hostogram for the same
df['AGE'].hist(bins=12)
plt.title("Age distributuion of the diabetic patients")
plt.xlabel("Age Range")
plt.ylabel("Count")
plt.show()
```

```python
df['BMI'].hist(bins=12)
plt.title("BMI distributuion of the diabetic patients")
plt.xlabel("BMI Range")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['BMI'],df['Y'], color = 'blue')
plt.xlabel("BMI ")
plt.ylabel("Disease progression")
plt.title("Scatter plot: BMI VS Disease progression")
plt.show()
```

```python
plt.scatter(df['S5'],df['Y'], color = 'blue')
plt.xlabel("S5_Blood Type ")
plt.ylabel("Disease progression")
plt.title("Scatter plot: S5_Blood Type VS Disease progression")
plt.show()
```
